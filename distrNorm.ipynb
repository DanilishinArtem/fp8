```python
import torch
import torch.distributions as dist
import matplotlib.pyplot as plt
```

```python
group_size = 512
normal_tensor = torch.normal(mean=0, std=1e-9, size=(16384, 16384)) * 1e-8
plt.hist(normal_tensor.view(-1), bins=50, density=True)
plt.show()
```

```python
group_size = 16384
normal_tensor = torch.normal(mean=0, std=1e-9, size=(16384, 16384)) * 1e-8
source_size = normal_tensor.numel()
normal_tensor = normal_tensor.view(group_size, -1).sum(dim=(1,))
plt.hist(normal_tensor, bins=50, density=True)
plt.show()
delta = pow(source_size / group_size, 1/2)
print('delta = {}'.format(delta))
```
