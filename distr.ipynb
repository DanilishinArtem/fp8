```python
import torch
import torch.distributions as dist
import matplotlib.pyplot as plt
```

```python
normal_tensor = torch.normal(mean=0.0, std=1.0, size=(1, 1000000)).view(-1)
plt.hist(normal_tensor, bins=50, density=True)
plt.show()
```

```python
import torch
import torch.distributions as dist
exp_dist = dist.Exponential(rate=0.001)
sample = exp_dist.sample((1000000,))
plt.hist(sample, bins=200, density=True)
plt.show()
```

```python
gourp_size = 100000
transformed = sample.view(gourp_size, -1).amax(dim=(1,))

plt.hist(transformed, bins=200, density=True)
plt.show()
```
